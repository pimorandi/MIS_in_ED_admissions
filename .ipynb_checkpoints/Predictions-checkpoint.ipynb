```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from utils import colorCode_performance, read_pkl, write_pkl, plot_roc

from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression

from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

import tensorflow as tf
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

from gensim.models import Word2Vec

from jhn_ai.jhn_ai.evaluation import supervised_crossvalidation
```

```python
df = read_pkl('./results/processed_df.pkl')
df.head()
```

```python
X = np.vstack(df.Embedding.values)
y = df.Target.values
```

```python
df_w = read_pkl('./word_similarity/stroke_preprocessed_df_RETURN_v3.pkl')

XX = np.vstack(df_w.DES_EMBED)
yy = df_w.TARGET.values

T = XX[yy==1]
P = T.mean(axis=0)

w2v = Word2Vec.load('./word_similarity/w2v_stroke.mdl')

pred_words = w2v.wv.most_similar(positive=[P], topn=17)

w = [t[0] for t in pred_words if t[0][:3]!='dev']
d = [t[1] for t in pred_words if t[0][:3]!='dev']

f, ax = plt.subplots(figsize=(8,5))
_ = ax.barh(y=w, width=d)
_ = ax.set_title("Top 15 most important Words", fontsize=12)
_ = ax.set_xlabel("[cosine similarity]")

plt.tight_layout()
```

# Predictions

## LightGBM

```python
args = dict(
    max_depth=Integer(5,50),
    learning_rate=Real(1e-5,1e-2, prior='log-uniform'),
    n_estimators=Integer(100,5_000),
    num_leaves=Integer(2,50),
    reg_lambda=Real(1e-5, 1e+5, prior='log-uniform'),
    reg_alpha=Real(1e-5, 1e+5, prior='log-uniform'),
)

lgb = LGBMClassifier(
    class_weight='balanced',
    random_state=71,
    n_jobs=-1
)
cv = StratifiedKFold(n_splits=5)
cq = StratifiedKFold(n_splits=5)
model = BayesSearchCV(
    lgb, 
    args,
    n_iter=100,
    n_points=10,
    cv=cq,
    scoring='f1_macro',
)


lgb_r = supervised_crossvalidation(
    model,
    X, y,
    cv=cv,
    disable_progressbar=True
)

lgb_r[0].groupby('metric').agg(['mean', 'std'])
```

```python
ax = plot_roc(lgb_r, figsize=(8,6))
```

```python
colorCode_performance(df, lgb_r)
```

## Logistic Regression

```python
from sklearn.calibration import CalibratedClassifierCV
from skopt.space import Real, Categorical, Integer
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(
    penalty='elasticnet',
    class_weight='balanced',
    solver='saga',
    random_state=71,
    n_jobs=-1
)

args = dict(
    C=Real(1e-5,1e+5, prior='log-uniform'),
    l1_ratio=Real(1e-7,1e+0, prior='log-uniform'),
)

cv = StratifiedKFold(n_splits=5)
cq = StratifiedKFold(n_splits=5)
model = BayesSearchCV(
    lr, 
    args,
    n_iter=100,
    n_points=10,
    cv=cq,
    scoring='f1_macro',
)


lr_r = supervised_crossvalidation(
    model,
    X, y,
    cv=cv, 
    disable_progressbar=True
)

lr_r[0].groupby('metric').agg(['mean', 'std'])
```

```python
ax = plot_roc(lr_r, figsize=(8,6))
```

```python
colorCode_performance(df, lr_r)
```

## Feed Forward Neural Network

```python
tf.random.set_seed(71)

class CustomKerasClassifier(KerasClassifier):
    def predict(self, X):
        return super().predict(X).flatten()
        

def build_function():
    clf = Sequential()
    clf.add(Dense(6, activation=tf.keras.layers.LeakyReLU(), input_dim=300))
    clf.add(Dropout(0.2))
    clf.add(Dense(1, activation='sigmoid'))
    clf.compile(loss='binary_crossentropy', optimizer='adam')
    return clf

nn = CustomKerasClassifier(build_function, epochs=20, batch_size=128, verbose=0)

cv = StratifiedKFold(n_splits=5)

neg, pos = np.bincount(y)
total = neg + pos
w0 = (1 / neg) * (total / 2.0)
w1 = (1 / pos) * (total / 2.0)

nn_r = supervised_crossvalidation(
    nn,
    X, y,
    cv=cv,
    fit_kwargs=dict(class_weight={0: w0, 1: w1}),
    disable_progressbar=True
)

nn_r[0].groupby('metric').agg(['mean', 'std'])
```

```python
ax = plot_roc(nn_r, figsize=(8,6))
```

```python
colorCode_performance(df, nn_r)
```
